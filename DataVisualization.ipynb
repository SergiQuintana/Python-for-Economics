## Figure
A `Figure` is the object in charge of representing the visualization. It is the graphic container. To create this object we will use the function `plt.figure()`. This will create an empty object, which we will be able to add information to. 

```python
plt.figure()
```

### Parts of the Figure

The main parts of a Matplotlib `Figure` are the following. Notice that the library will allow as to customize all of those features. 



Let's now review some of the main graphs we can do. 

## Types of Graphs
### Line Charts

It is the standard graph representing a line that connects the different data points. To create a line chart we will use the command `plt.plot()`.


```python
import numpy as np
import pandas as pd
fig = plt.figure()  # create the figure object.
plt.plot(pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x)))
```

Notice that we are only ploting over the `y-axis`. The `x-axis` is just the index of the elements. We can change that by including another list of data. 

```python
fig = plt.figure()  # create the figure object.
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x)))
```

Notice that we can plot multiple lines in the same `Figure`. 

```python
fig = plt.figure()  # create the figure object.
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x)))
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.linspace(10,0,10)).apply(lambda x: x/(10+x)))
```

## Bar Charts

We can use bar charts to represent categorical data. The length of the bar is proportional to the value of the variable. 


```python
x = np.arange(0,10)
y = np.random.randint(0,3,10)

fig = plt.plot()
plt.bar(x,y)
```

We can also do it with categorical data. 

```python
names = ['Maria','Jhon','Peter','Alba']
grade = [10,4,5,7]

fig = plt.plot()
plt.bar(names,grade)
```

### Bar graphs with multiple variables
We can also create bar graphs of two variables for the same category. For example, we could plot the grades for two courses.

```python
maths = [10,4,5,7]
econometrics = [9,7,8,5]

width = 0.35  # it is the width of the bars
location = np.arange(len(names))  # location of the names on the x-axis. 

fig,ax = plt.subplots()   # we use the subplot!
ax.bar(location -width/2,maths,width)
ax.bar(location+width/2,econometrics,width)
ax.set_xticks(location,names)
```

### Stack bars

We can also stack mutliple bars on top of each other. We use the argument `bottom` to specify that we are plotting a bar graph on top of another.


```python
fig = plt.plot()
plt.bar(names,maths)
plt.bar(names,econometrics,bottom=maths)
```

## Histograms
Histograms are used to have a sense of how the data is distributed in our dataset.  They use a bar representation to show the data belonging to each range. Back to our Pokemon data we can use some examples: 

```python
df = pd.read_csv('Pokemon.csv')
fig = plt.plot()
plt.hist(df.Attack)
```

### Changing the amount of bins
We can choose the amount of bins we want. However, this can make our Histogram less informative, so one has to be cautious. 

```python
fig = plt.plot()
plt.hist(df.Attack,bins=4)
```

Furthermore, we can create bins of unequal sizes by inputinz the bins bounds. 

```python
fig = plt.plot()
plt.hist(df.Attack,bins=[4,25,60,63,100,200])
```

### Multiple Histograms
We can inlcude multiple histograms on the same chart. Furthermore, to improve the visualization we can reduce the opacity using `alpha`. 

```python
fig = plt.plot()
plt.hist(df.Attack,alpha=0.6)
plt.hist(df.HP,alpha=0.4)
```

## Scatter plot

We just need to use the function `scatter()`. 

```python
plt.figure()
plt.scatter(df.Attack,df.HP)
```

### Scatter plot with multiple variables
We can also produce a `Figure` showing two scatter plots at the same time. 

```python
plt.figure()
plt.scatter(df.Attack,df.HP)
plt.scatter(df.Attack,df.Speed)
```

## Charts Customization

We will now learn to customize our `Figures`. You will see that at some point we will define the `Figure` as `fig,ax = plt.subplots()` instead of `fig = plt.figure()`. This is a technicality that will improve our custimization capabilites. If you would like to know more about it, visit [here](https://towardsdatascience.com/clearing-the-confusion-once-and-for-all-fig-ax-plt-subplots-b122bb7783ca).

#### Figure Size

We can modify the size of the overall `Figure`.

```python
fig, ax = plt.subplots(figsize=(10,4))
```

#### Include labels

We can include labels to the axis of the `Figure` using the functions `xlabel` and `ylabel`. 

```python
fig = plt.figure()  # create the figure object.
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x)))
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.linspace(10,0,10)).apply(lambda x: x/(10+x)))
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
 # When working with fig, ax we can use ax.set_xlabel
    
fig,ax = plt.subplots(figsize=(3,3))  # create the figure object.
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x)))
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.linspace(10,0,10)).apply(lambda x: x/(10+x)))
ax.set_xlabel('Age')
ax.set_ylabel('Income')
```

#### Include a title

To inlcude a title we just need to use the method `plt.title()`. 

```python
fig = plt.figure()  # create the figure object.
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x)))
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.linspace(10,0,10)).apply(lambda x: x/(10+x)))
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Lifecycle Income Comparison')
```

#### Include a legend

Finally, we can also label our lines and include a legend to know the meaning of each line.

```python
fig = plt.figure()  # create the figure object.
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x)),label='Jhon')
plt.plot(np.linspace(3,50,10),pd.DataFrame(np.linspace(10,0,10)).apply(lambda x: x/(10+x)),label='Peter')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Lifecycle Income Comparison')
plt.legend()
```

### Line and Marker styles

Since its a customizable `Figure` we can change the [linstyle](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html) and the [marker](https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html) style, just like with Matlab. 

```python
# Define our lines outside:
a = pd.DataFrame(np.arange(0,10)).apply(lambda x: x/(10+x))
b = pd.DataFrame(np.linspace(10,0,10)).apply(lambda x: x/(10+x))

# Figure:

fig = plt.figure()  # create the figure object.
plt.plot(np.linspace(3,50,10),a,label='Jhon',linestyle = 'dashed',marker='o')
plt.plot(np.linspace(3,50,10),b,label='Peter')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Lifecycle Income Comparison')
plt.legend()
```

#### Limit the axis
We can include limitations in the range of values that we allow in our axis. We will use `xlim` and `ylim`.

```python
fig, ax = plt.subplots()
plt.scatter(df.Attack,df.HP)
plt.xlim([50,90])
```

#### Include a Grid
We can include a grid in the background of our `Figure` to improve its layout. 

```python
fig,ax = plt.subplots()
plt.scatter(df.Attack,df.HP)
ax.grid()
```

We can modify the color and the linestyle of the grid.

```python
fig,ax = plt.subplots()
plt.scatter(df.Attack,df.HP)
ax.grid(color='r',linestyle='--')
```

Some of the options are: 
- `color = ''` Examples: 
    - Red: `'r'` 
    - Blue: `'b'`
    - Green: `'g'`
    - Violet: `'violet'`

- `linestyle= ''` Examples:
    - `'-'` or `'solid'`
    - `'--'` or `'dashed'`
    - `'-.'` or `'dashdot'`
    - `':'` or `'dotted'`


## Subplots

We can also include more than one plot inside one `Figure`.  When creating our subplot object we can specify the number of rows and columns our `Figure` will have. Each entry corresponds to one subplot. 

```python
fig, (ax1,ax2) =  plt.subplots(1,2)  #   1 row and 2 columns
```

Notice that now we have two `ax` objects. One for each of our subplots. If we do not specify, then the `ax` object will become a list and we can acccess each of the axis. 

```python
fig, ax = plt.subplots(1,2)
ax[0].scatter(df.Attack,df.HP)
ax[1].scatter(df.Attack,df.Speed)
```

We can now set different axis names and titles for each of the `subplots`. 

```python
fig, ax = plt.subplots(1,2)
ax[0].scatter(df.Attack,df.HP)
ax[1].scatter(df.Attack,df.Speed)

ax[0].set_xlabel('Hello')
ax[0].set_ylabel('Goodbye')

ax[1].set_xlabel('Econometrics')
ax[1].set_ylabel('Maths')
fig.text(0.5,0.04, "Some text", ha="center", va="center")
```

## Save Figures

To save figures we just need to use `plt.savefig()`. 

```python
fig,ax = plt.subplots()
plt.scatter(df.Attack,df.HP)
ax.grid(color='r',linestyle='--')
plt.savefig('test.png')
```

# Seaborn

Seabron is a data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics with easier syntaxt than matplotlib. It is easily integrated with Pandas, which makes our life easier.  To install `seaborn` just type :

```python
!pip install seaborn
```

 For convention seaborn is imported as `sns`. We still need to import `matplotlib` when working with seaborn. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Line Graph

To create a line graph we use `lineplot()`.

```python
fig = sns.lineplot(x='Attack',y='HP',data=df)
fig.set_title('Line Graph')
```

We can create a line graph for each subgroup using the option `hue`.  Also notice that the previous graph displays a confidence interval, we can turn it off with `ci=None`. 

```python
fig = sns.lineplot(x='Attack',y='HP',data=df,ci=None,hue='Legendary')
```

## Scatter Plots


```python
fig = sns.scatterplot(x='Attack',y='HP',data=df,ci=None,hue='Legendary')
fig.set_title('Scatter Plot of Attack and HP')
```

## Histograms

```python
fig = sns.histplot(x='Attack',data=df,hue='Legendary',kde=True)  # option kde shows the kernel density
```

## Swarmplot
It draws a categorical scatterplot with points adjusted to be non-overlapped. It allows to give a sense of the dispersion and the distribution of a variable.


```python
fig = sns.swarmplot(x='Legendary',y='Attack',data=df)
```

```python
fig = sns.swarmplot(x='Legendary',y='Attack',data=df,hue='Type 1')
```

## Heatmap

It allows to plot data as a color matrix.


```python
glue = sns.load_dataset("glue").pivot("Model", "Task", "Score")
sns.heatmap(glue)
```
