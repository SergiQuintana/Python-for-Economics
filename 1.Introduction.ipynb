## Python technical part (or how python runs programs)

The first important concept is the Python interpreter. It is the software in charge of reading and executing the instructions of our programs. In order to run a Python code we must have the interpreter installed in our computer.  See installation HERE. How does it work? The user would type code into text files and those files would run throgh the interpreter. But what is the interpreter doing? Let's take a look!

The first thing that Python does when we execute a program is to translate the instructions of our file into byte code. Basically, the interpreter translates the Python language into a lower-level lenguage and creates a compiled file that Python will store. This will save us execution time, because if in the future we execute the code again, Python will use the compiled version and skip the compilation. 

Once the code is translated into byte code it is taken into the runtime engine of Python, known as the Python Virtual Machine (PVM). The PVM iterates over the byte code instructions one by one to carry the instructions and produce the desired output.

<div class="alert alert-block alert-info"> <b>TIP!</b> Note about Python variable creation. A variable in Python is created when in the code we first asign a value to it. Variables are just names that reference particular objects. For example, when we create a variable <b> a = 3 </b> Python is doing the following: 
<ol>
    <li> Create an object to represent the value of 3.</li>
    <li> Create the variable <b> a </b> if it does not exist yet </li>
    <li> Link the variable <b>a</b> to the new object 3. </li>
    </ol>
It is very important to understand that variables always link to objects and never to other variables. In that sense, links from variables to objects are called references. 
</div> 

On the previous box Python variable creation is discussed. The reason to do that is to understand what happens when to variables point to the same object, and to justify the need of methods like `copy()`. At first sight the reader might think that the copy() method is useless since we can just create a new variable equal to the past one. 


````python
# Example
text = ['book','laptop','green','book','orange']
text2 = text 
print(text2)

````

However, by asigning `text2 = text` all we are doing is referencing the new variable `text2` to the already existing object `['book','laptop','green','book','orange']`. What are the implications of this? If we produce any change in the variable `text2` it will also change the variable `text` since they are pointing to the same object. 

||
|:--:|
|*Python Shared Reference. Source: Fig 6.2. from Learing Python by Mark Lutz.*|


````python
text2[2] = 'House'  # replace the 3rd element of text2. 
print(text)         # The element also changes in text! 

# For this reason we use copy()
text = ['book','laptop','green','book','orange']
text2 = text.copy()

text2[2] = 'House'

print(text)
print(text2)
````

By using the method `copy()` we are creating a new object in the memory of Python, and we can safely chaning withouht accidentally changing the initial object. 

#### Dictionaries
Python dictionaries are knonw as *mappings*. Mappings are collections of objects that are stored by their *key* instead of by their relative position. Each *value* of a dictionary has an asociated *key*.  In a dictionary, both the *key* and the *value* can be of any type. 

````python
first_dicc  = {'name': 'Jhon', 'surname':'Smith'}  #first element key, second value

print(first_dicc)

mydict = {1:'house',23:'flat'}  #keys are numbers, values are strings
print(mydict) 

mydict = {2:'door','number':34} # mixt of types
print(mydict)

mydict = {'names':['Jhon','Angela'], 'age':[13,18]}   #we can even do a dict with a list
print(mydict)

# We can also create dictionaries using other ways:
bob1 = dict(name='Bob',job='dev',age=40)
print(bob1)

bob2 = dict(zip(['name','job','age'],['Bob','dev',40]))
print(bob2)
````

#### Methods and opperations 

We will now review the most common methods and opperations with dictionaries. Some of the methods are: 
<ul>
    <li> <b> Get </b>: Returns a value based on a key.</li>
    <li> <b> Keys</b>: Will return a list of all the keys in the dictionary.</li>
    <li> <b> Values </b>: Will return a list of all the values in the dictionary. </li>
    <li> <b> Items </b>: Will return each item in the dictionary as tuples on a list. </li>
    <li> <b> Update </b>: Will change the value of a key, or introduce a new key and value if the key did not exist before. </li>
    <li> <b> Pop </b>: Like with lists, it will eliminate the item with a specific key.  </li>
    <li> <b> Clear </b>: Empties the dictionary.</li>

````python
# GET
bob1 = dict(name='Bob',job='dev',age=40)
print(bob1.get('name'))

# KEYS
print(bob1.keys())

#VALUES
print(bob1.values())

#ITEMS
print(bob1.items())
````

````python
#UPDATE

bob1.update({'name':'Jhon'})    #change an already exisiting key
print(bob1)

bob1.update({'married':'Yes'}) # include a new item
print(bob1)

# the same can be done by indexing:

bob1['name'] = 'Albert'  # change an already existing key
print(bob1)

bob1['kids'] = 2    # include a new item
print(bob1)
````

````python
#POP
bob1.pop('name')   # delete one item
print(bob1)

# CLEAR
bob1.clear()
print(bob1)
````

### List Comprehension

The list comprehensions is an advanced operation which turns out  to be a powerful way to process list structures. It is a way of slicing lists based on logic operators. List comprehensions are very elegenat and their use is always encouraged. Let's see an example:

````python
# Example with a list of lists: 
M = [[1,2,3],[4,5,6],[7,8,9]]   # a list of lists (similar to a matrix)

# Suppose we would like to extract the first row. We can just do
print(M[0])

# However, what if we want to extract the second column?

print([row[1] for row in M])    # We can use a list comprehension!

````

List comperhensions derive from set notation; they are a way to build a new list by running an expression on each item in a sequence, one at a time, from left to right. List comperhensions are coded in squared brackets (to signal that they make a list) and are copmised of an expression and a looping construct that share a variable name (*row* in our example). We can have more complex examples:


````python
print([row[1] for row in M if row[1] % 2 == 0])   # filter out odd items
````

List comperhensions make new list of results, but they can be used to iterate over any iterable object. For instances in the next example we use list comprehensions to step over a hardcoded list of coordinates:

````python
diag = [M[i][i] for  i in range(3)]   # take the diagonal elements
print(diag)
````

````python
# Another example is to classify odd and even numbers in the range of 10

print([[num,'even'] if num % 2 != 0 else [num,'odd'] for num in range(1,11)])
````

We can also introduce two loops in a list comperhension. For example to eliminate one dimension from our list:

````python
# Back to the matrix M
M = [[1,2,3],[4,5,6],[7,8,9]]

print([num for row in M for num in row])   # we are eliminating one dimension!
````

#### Keywords

In Python, we can be more specific about what goes where when we call a function. Keyword arguments allow us to match by <b> name </b> , instead of by position. Let's see an example:

````python
# Keywords

def func4(a,b,c):
    print(a)
    print(b+c)
    return a+b+c

# Under positional arguments we would do:
func4(1,2,3)

# However, we can match according to the name:
 
func4(b=1,c=5,a=10)   # notice now the argument for the parameter a is provided in the third place, but it is not a problem!
````

````python
func4(3,c=10,b=2)
````

#### Class and Attributes
Tho create a `class` we just need to use the same syntaxt as with functions, but now using the `class` statement. Once a `class` is created, we would need to specify what are its attibutes. Those will be the properties and characteristics that will shape our class. To establish the attributes we will use the `__init__` method, which is known as the constructor method and is used to initialize objects state. Furthermore, to define the parameters of the class, we will use the `self` statement, that will allow to access the parameterst that belong to the class. The attributes defined in `__init__` are known as instance attributes.  Let's take a look:

````python
# Let's create a class:

class Person:                # define the class
    
    def __init__(self,name,age,city):  # create the method that will define its attributes
        
        self.name = name               # set the class characteristics
        self.age = age
        self.city = city
        
jhon = Person('Jhon',34,'Barcelona')   # create a new object called jhon with the class characteristics. 
jhon.name
````

````python
jhon.__dict__
````

In the previous exaple we have seen different things. First of all unlike with functions, in a class we do not define the attributes after their name (`class Person(name,age,city)`). We do it inside the constructor `__init__(self,name,age,city)`. Also, notice that when callig a class we create a new object, in this case `jhon` who has the properties of the parameters we introduce. If we would like to access those characteristics we can just type for example `jhon.name`. Notice that once we created the `jhon` object we can still change its parameter values:

````python
jhon.age= 45
jhon.age
````

The last important concept to be introduced inside our `class` are methods. Methods are functions inside a class, which are coded as normal `defs` and they support everything we've learned about functions. The peculiarity is that the first argument will recieve the `self` argument which will automatically pass all the variables that we have defined in `__init()__`. We can also include other variables that are method specific.

````python
class Person:                # define the class
    
    def __init__(self,name,age,city):  # create the method that will define its attributes
        
        self.name = name               # set the class characteristics
        self.age = age
        self.city = city
        
    def sentence(self):
        print(f"Hi! My name is {self.name}, I'm {self.age} and I live in {self.city}")
        
    def check_city(self,newcity):
        
        if self.city == newcity:
            print('Wow! We are from the same city')
        else:
            print(f"Nice, I'm from {newcity}")
            
jhon = Person('Jhon',34,'Barcelona')    # we just need to provide the instance atributes

jhon.sentence()
jhon.check_city('London')
````

Notice we could create another object with other instance attributes. 

````python
marta = Person('Marta',25,'Madrid')
marta.sentence()
````

#### Class Inheritance

Besides serving as factories for generating multiple instance objects, classes also allow us to make changes by introducing new components (called subclasses) instad of changing existing components in place.  Python allows classes to inherit from other classes, opening the door to coding hierarchies of classes. The classes inherit from superclasses. Classes inherit all of the attribute names defined in their superclasess, and can also have attributes and methods of their own. Let's create a subclass from `Persons`. 

````python
# simple example
class Spanish(Person):
    pass

marta = Spanish('Marta',25,'Madrid')
marta.name
````

The previous subclass is a very simple example, that adds nothing to the superclass `Person`. However, we could add new methods and even new instance attributes. 


````python
class Spanish(Person):
    
    def __init__(self,name,age,city,food):
        
        super().__init__(name,age,city)  # we are inheritng the isntance attributes from Person
        self.food = food
        
    def favfood(self):
        print(f'My favourite food is {self.food}')
        
pedro = Spanish('Pedro',34,'Madrid','paella')
pedro.sentence()          # we can call methods from Person
pedro.favfood()           # and also we can call methods from Spanish . 
````
